```python
import pandas as pd
data=pd.read_csv('StudentPerformance (1).csv')
df=pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
#after finding the missing values we need to drop
df.dropna()
```

```python
df.fillna(80)
```

```python
#to check duplicates
df.duplicated()
```

```python
#to remove duplicates
df.drop_duplicates()
```

```python
df['Hours Studied'].value_counts()
```

```python
df['Hours Studied'].mean()
```

```python
df.groupby('Hours Studied')['Sleep Hours'].sum()
```

```python
df['Previous Scores'].max()
```

```python
df['Previous Scores'].value_counts()
df['Sleep Hours'].value_counts()
df['Previous Scores'].max()
df['Sleep Hours'].min()
```

```python
#filtering data
df[df['Previous Scores']>60]
```

```python
#multiple conditions means compound conditions
df[(df['Previous Scores']>60)&(df['Sleep Hours']>5)]
```

```python
df.to_csv('StudentPerformance.csv')
```

```python
#saving cleaned data set
df.to_csv('StudentPerformance.csv',index='False')
```
